```python
import glob
import os
from PIL import Image
import pandas as pd


image_dir = '/kaggle/input/datasets/organizations/nih-chest-xrays/sample/sample/sample/images'
output_path = '/kaggle/working/nih_meta.csv'


rows = []
for image_path in glob.glob(os.path.join(image_dir, '*.png')):
    image_id = os.path.splitext(os.path.basename(image_path))[0]
    with Image.open(image_path) as img:
        width, height = img.size  # PIL: (width, height); lazy load, doesn't decode pixels
    rows.append({'image_id': image_id, 'dim0': height, 'dim1': width})

test_meta = pd.DataFrame(rows, columns=['image_id', 'dim0', 'dim1'])
test_meta.to_csv(output_path, index=False)
print(f'Wrote {len(test_meta)} rows to {output_path}')
```
